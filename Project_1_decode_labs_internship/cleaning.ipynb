```python
import pandas as pd
import numpy as np
df = pd.read_excel("Dataset for Data Analytics.xlsx")
print(df.head())
```

```python
initial_len = len(df)
df = df.drop_duplicates()
print(f"Dropped {initial_len - len(df)} exact duplicate rows.")
```

```python
df['CouponCode'] = df['CouponCode'].fillna('NONE')
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date']
```

```python
dups_order_id = df[df.duplicated(subset=['OrderID'], keep=False)]
print("No. of duplicate OrderID's",dups_order_id['OrderID'].unique())
```

```python
string_cols = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource', 'CouponCode']

df[string_cols] = df[string_cols].apply(lambda col: col.astype(str).str.strip())

print(df.dtypes)
```

```python
expected_total = (df['Quantity'] * df['UnitPrice']).round(2)

mismatches = df[(df['TotalPrice'].round(2) - expected_total).abs() > 0.05]

print(f"Total rows in dataset: {len(df)}")
print(f"Number of rows with mismatched TotalPrice: {len(mismatches)}")
if len(mismatches) > 0:
    print("\nSample of mismatched rows:")
    print(mismatches[['OrderID', 'Quantity', 'UnitPrice', 'TotalPrice']].head())
```

```python
print("Unique Products:")
print(df['Product'].unique())
print("\nUnique Payment Methods:")
print(df['PaymentMethod'].unique())
print("\nUnique Order Statuses:")
print(df['OrderStatus'].unique())
print("\nUnique Referral Sources:")
print(df['ReferralSource'].unique())
```

```python
print("Invalid/Missing Dates:", df['Date'].isna().sum())

df.to_excel("Cleaned_Dataset_for_Data_Analytics.xlsx", index=False)
print("\nSuccessfully saved the cleaned dataset to 'Cleaned_Dataset_for_Data_Analytics.xlsx'!")
```
